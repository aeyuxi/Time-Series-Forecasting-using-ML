Several ideas come to my mind to explore in detail the data. The previous plots iustre how the sales behave in all stores amd items. Carefull separation will take place later on, for now, I'd like to explore the idea of predicting sales in the future. We can see that the graph has some tendency and a distinguished pattern, let's start with that. 

Let's take  and normalize it the dataset

```python
dataset = resample_data_months['sales']
dataset_list = list(dataset)
dataset_norm = [float(i)/max(dataset_list) for i in dataset_list]
dataset_array = np.array(dataset_norm)
dataset_array = np.reshape(dataset_array,[len(dataset),1])
```

Let's separate the data in trainning and testing 

```python
dataset_array_trainning = dataset_array[0:40]
dataset_array_testing = dataset_array[41:60]
```

Let's define NN model that will be used for trainning. Remember that the idea is  to use the data to find the time series model. Here is where the power of ML relies. 

```python
try:
    n_output = 1
    n_features = 1
    #Let's include 10 hidden layers
    layer_dim_list = [n_features, 10, n_output]
    
    X = tf.placeholder( dtype=tf.float32,
                        shape = [None, n_features], 
                        name = 'input')
    
    Y = tf.placeholder( dtype=tf.float32,
                       shape = [n_output], 
                       name = 'output')
    
    current_input = X        
    Ws_list =[]
    
    for i, layer_i in enumerate(layer_dim_list):    
        with tf.variable_scope("percpetron/{}".format( int( layer_i ) ) ):
            n_input = int( layer_dim_list[i] )
            if i + 1 < len( layer_dim_list ):
                n_output = int(layer_dim_list[i + 1])
            else:
                break
            
            W = tf.get_variable( name = 'W',
                                 dtype=tf.float32,
                                 shape = [n_input, n_output],
                                 initializer=tf.random_normal_initializer(mean=0.0, stddev=0.01))
            
            b = tf.get_variable( name = 'b',
                                 shape = [n_output],
                                 dtype=tf.float32,
                                 initializer=tf.constant_initializer( 0.01 ) )
            
            
            h = tf.nn.bias_add( name='h',
                                value=tf.matmul(current_input, W),
                                bias=b)
        
            current_input = tf.nn.relu( h ) 
        
            Ws_list.append(W)
            
    pred = current_input    
    cost = tf.reduce_mean( tf.square( h - Y ) )
    optimizer = tf.train.GradientDescentOptimizer(0.2).minimize(cost)
    init = tf.global_variables_initializer( )
    cost_list = []
    with tf.Session( ) as sess:
        sess.run( init )
        n_iterations = 40
        avg_cost_pred = 0
        for it_i in range( n_iterations ):
            cost_pred_ac = 0
            for sale in dataset_array_trainning:
                _ , cost_pred, val_pred = sess.run( [optimizer, cost, pred],feed_dict={X: np.reshape(sale,[1,1]), Y : sale }) 
                cost_pred_ac += cost_pred    
            avg_cost_pred =  cost_pred_ac / len( dataset_array_trainning )
            cost_list.append(cost_pred)    
            print("cost  :",avg_cost_pred)
        predictions = current_input.eval( feed_dict = {X: dataset_array_testing} )
               
        fig_result, axs_result = plt.subplots(2, 1, constrained_layout=True)
        axs_result[0].plot(cost_list)
        axs_result[0].set_xlabel('Epoch')
        axs_result[0].set_ylabel('Cost')
        
        axs_result[1].plot(predictions, color = 'green')
        axs_result[1].plot(dataset_array[41:60], color = 'red')
        axs_result[1].set_xlabel('date')
        axs_result[1].set_ylabel('Normalized Sales')
        
        
except tf.errors.InvalidArgumentError as e:
   print( e )

finally:
    tf.reset_default_graph(  )
```
